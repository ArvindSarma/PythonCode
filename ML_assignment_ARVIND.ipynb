```python
import os
import numpy as np
import pandas as pd
import math
import matplotlib.pylab as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
# Load data
path = "./churn2.csv"
df = pd.read_csv(path)[["COLLEGE", "INCOME", "OVERAGE", "LEFTOVER", "HOUSE","HANDSET_PRICE","OVER_15MINS_CALLS_PER_MONTH","AVERAGE_CALL_DURATION","REPORTED_SATISFACTION","REPORTED_USAGE_LEVEL","CONSIDERING_CHANGE_OF_PLAN","LEAVE"]].dropna()
# Take a look at the data
df #DATAFRAME First 5 objects
```

```python
df["COLLEGE2"] = (df.COLLEGE == "one").astype(int)
df
```

```python
df.dtypes
```

```python
df.REPORTED_SATISFACTION = df.REPORTED_SATISFACTION.astype('str')
df.REPORTED_USAGE_LEVEL = df.REPORTED_USAGE_LEVEL.astype('str')
df.CONSIDERING_CHANGE_OF_PLAN = df.CONSIDERING_CHANGE_OF_PLAN.astype('str')
df.dtypes
```

```python
df["LEAVE2"] = (df.LEAVE == "STAY").astype(int)
df.head(5)
```

```python
# Names of different columns
predictor_cols = ["INCOME","HOUSE","COLLEGE2"]
target_col = "LEAVE2"
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df[predictor_cols],df[target_col],test_size = 0.25,random_state = 0)
```

```python
print("X_train shape: {}".format(X_train.shape))
print("X_test shape: {}".format(X_test.shape))
print("y_train shape: {}".format(y_train.shape))
print("y_test shape: {}".format(y_test.shape))
```

```python
X_test
```

```python
y_test
```

```python
from sklearn.tree import DecisionTreeClassifier
# Let's define the model (tree)
decision_tree = DecisionTreeClassifier(max_depth=6, criterion="entropy",max_leaf_nodes = 12, min_samples_leaf = 1)
# Let's tell the model what is the data
decision_tree.fit(X_train, y_train)
```

```python
from sklearn.model_selection import cross_val_score
dt_accuracy = cross_val_score(decision_tree, df[predictor_cols], df[target_col], cv=10)
```

```python
print(f"Accuracy for 10 folds: {dt_accuracy}")
print("Decision Tree Average accuracy across the 10 folds: {:.2f}".format(dt_accuracy.mean()))
print("Decision Tree Standard Deviation of accuracy across the 10 folds: {:.2f}".format(dt_accuracy.std()))
```

```python
from sklearn.linear_model import LogisticRegression
lr_model = LogisticRegression()
lr_model
```

```python
from sklearn.model_selection import cross_val_score
lr_accuracy = cross_val_score(lr_model, df[predictor_cols], df[target_col], cv=10)
lr_accuracy
```

```python
print(f"Accuracy for 10 folds: {lr_accuracy}")
print("Logistic Regression Average accuracy across the 10 folds: {:.2f}".format(lr_accuracy.mean()))
print("Logistic Regression Standard Deviation of accuracy across the 10 folds: {:.2f}".format(lr_accuracy.std()))
```
